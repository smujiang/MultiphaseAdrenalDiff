```python
import pandas as pd
import os

from config import DATA_ROOT_DIR
original_text_data_dir = os.path.join(DATA_ROOT_DIR, "original_data/data_pixels")
processed_data_root_dir = os.path.join(DATA_ROOT_DIR, "processed_data")
updated_adrenal_raw_save_dir = os.path.join(processed_data_root_dir, "pixel_data")

csv_file_list = os.listdir(original_text_data_dir)
for csv_file in csv_file_list:
    if csv_file.endswith(".txt"):
        csv_file_path = os.path.join(original_text_data_dir, csv_file)
        print(f"Processing {csv_file_path}...")
        
        # Load the CSV file
        df = pd.read_csv(csv_file_path,sep='\t')
        
        # Remove all spaces from the entire DataFrame
        df.columns = df.columns.str.replace(' ', '')
        df = df.map(lambda x: x.replace(' ', '') if isinstance(x, str) else x)
        
        # Save the updated DataFrame back to a CSV file
        updated_csv_file_path = os.path.join(updated_adrenal_raw_save_dir, csv_file.replace(".txt", ".csv"))
        df.to_csv(updated_csv_file_path, index=False)
```
